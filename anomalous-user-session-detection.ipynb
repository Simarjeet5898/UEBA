```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm import tqdm

from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.cluster import DBSCAN
from sklearn.metrics import mean_squared_error

from keras.models import Model
from keras.layers import Input, Dense
```

```python
# Load datasets from Kaggle input path
train_path = "/kaggle/input/anomalous-user-session-tracking/user_session_tracking_synthetic_training.csv"
# test_path = "/kaggle/input/anomalous-user-session-tracking/user_session_tracking_synthetic_testing.csv"
test_path = "/kaggle/input/user-session-tracking-labelled/user_session_tracking_synthetic_testing_labeled.csv"


# Parse login/logout timestamps
train_df = pd.read_csv(train_path, parse_dates=["login_time", "logout_time"])
test_df = pd.read_csv(test_path, parse_dates=["login_time", "logout_time"])

print("Training shape:", train_df.shape)
print("Testing shape:", test_df.shape)
train_df.head()
```

```python
# Drop columns not needed for clustering
# cols_to_drop = ["mac_addresses", "hostname","active_mac"]
# train_df.drop(columns=cols_to_drop, inplace=True)
# test_df.drop(columns=cols_to_drop, inplace=True)
cols_to_drop = ["mac_addresses", "hostname", "active_mac", "is_anomaly"]  # don't use label in clustering
train_df.drop(columns=cols_to_drop[:-1], inplace=True)  # train_df has no is_anomaly
test_labels = test_df["is_anomaly"].copy()  # Save ground truth labels for evaluation
test_df.drop(columns=cols_to_drop, inplace=True)

# Extract time-based features
for df in [train_df, test_df]:
    df["login_hour"] = df["login_time"].dt.hour
    df["logout_hour"] = df["logout_time"].dt.hour
    df["day_of_week"] = df["login_time"].dt.weekday  # 0 = Monday, 6 = Sunday

print("Columns after processing:", train_df.columns.tolist())
train_df.head()
```

```python
# Encode categorical columns
categorical_cols = ["username", "auth_type", "remote_ip", "lan_ip", "source_os"]
label_encoders = {}

for col in categorical_cols:
    le = LabelEncoder()
    train_df[col] = le.fit_transform(train_df[col])
    label_encoders[col] = le

    # Map same labels to test set (unseen items → -1)
    mapping = dict(zip(le.classes_, le.transform(le.classes_)))
    test_df[col] = test_df[col].map(mapping).fillna(-1).astype(int)

print("Categorical encoding completed.")
train_df.head()
```

```python
# Select features for DBSCAN clustering
features = [
    "login_hour",
    "logout_hour",
    "day_of_week",
    "session_duration_seconds",
    "auth_type",
    "remote_ip",
    "lan_ip"
]

X_train = train_df[features].copy()
X_test = test_df[features].copy()

print("Selected features:\n", features)
print("\nSample of unscaled training data:\n", X_train.head())

# Normalize using StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# print("\nScaled training data sample (first 5 rows):\n", X_train_scaled[:5])
print("\n Features selected and scaled. Ready for Training.")
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from collections import Counter

# Step 1: Reduce dimensions of the scaled training data to 2D using PCA
# This helps us visualize high-dimensional clustering results
pca = PCA(n_components=2)
X_train_pca = pca.fit_transform(X_train_scaled)

# Step 2: Define a list of DBSCAN parameter combinations to try
# Each tuple contains (eps, min_samples)
param_grid = [
    (1.5, 5),
    (1.2, 5),
    (1.0, 5),
    (1.5, 10),
    (1.2, 10),
    (1.0, 10),
]

# Step 3: Create a grid of subplots to compare clustering results
plt.figure(figsize=(15, 10))

# Step 4: Loop through each parameter combination
for i, (eps, min_samples) in enumerate(param_grid, start=1):
    # Initialize and fit DBSCAN
    db = DBSCAN(eps=eps, min_samples=min_samples)
    labels = db.fit_predict(X_train_scaled)

    # Print the cluster summary for this combination
    print(f"DBSCAN eps={eps}, min_samples={min_samples}")
    print("Label counts:", Counter(labels))           # Shows how many points in each cluster
    print("Anomalies detected:", list(labels).count(-1))  # -1 means anomaly

    # Count number of clusters (excluding -1)
    n_clusters = len(set(labels)) - (1 if -1 in labels else 0)
    print("Number of clusters formed:", n_clusters, "\n")

    # Step 5: Plot the results in the subplot
    plt.subplot(2, 3, i)
    plt.scatter(
        X_train_pca[:, 0],        # PCA Component 1
        X_train_pca[:, 1],        # PCA Component 2
        c=labels,                 # Color points by cluster label
        cmap="viridis",
        s=4,
        alpha=0.6
    )
    plt.title(f"eps={eps}, min_samples={min_samples}")
    plt.xlabel("PCA 1")
    plt.ylabel("PCA 2")

# Step 6: Adjust layout and display all subplots
plt.tight_layout()
plt.show()
```

```python
print("Applying DBSCAN on test data (eps=1.5, min_samples=10)...")

# Fit DBSCAN with selected best config
test_dbscan = DBSCAN(eps=1.5, min_samples=10)
predicted_labels = test_dbscan.fit_predict(X_test_scaled)

# Assign predicted cluster labels and anomaly flag
test_df["cluster_label"] = predicted_labels
predicted_anomalies = (predicted_labels == -1).astype(int)
test_df["is_anomaly"] = predicted_anomalies  # Optional: only for inspection or plotting

# Summary
from collections import Counter
print("DBSCAN on test data completed.")
print("Cluster label counts:", Counter(predicted_labels))
print("Total anomalies detected in test data:", predicted_anomalies.sum())
```
