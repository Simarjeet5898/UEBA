```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm import tqdm

from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import mean_squared_error

import hdbscan 
```

```python
# Load datasets from updated Kaggle input path
train_path = "/kaggle/input/usersessiontrackingdataset/user_session_tracking_synthetic_training.csv"
test_path = "/kaggle/input/usersessiontrackingdataset/user_session_tracking_synthetic_testing_labeled.csv"

# Parse login/logout timestamps
train_df = pd.read_csv(train_path, parse_dates=["login_time", "logout_time"])
test_df = pd.read_csv(test_path, parse_dates=["login_time", "logout_time"])

print("Training shape:", train_df.shape)
print("Testing shape:", test_df.shape)
train_df.head()
```

```python
# Drop columns not needed for clustering
cols_to_drop = ["mac_addresses", "hostname", "active_mac", "is_anomaly"]  # don't use label in clustering
train_df.drop(columns=cols_to_drop[:-1], inplace=True)  # train_df has no is_anomaly
test_labels = test_df["is_anomaly"].copy()  # Save ground truth labels for evaluation
test_df.drop(columns=cols_to_drop, inplace=True)

# Extract time-based features
for df in [train_df, test_df]:
    df["login_hour"] = df["login_time"].dt.hour
    df["logout_hour"] = df["logout_time"].dt.hour
    df["day_of_week"] = df["login_time"].dt.weekday  # 0 = Monday, 6 = Sunday

print("Columns after processing:", train_df.columns.tolist())
train_df.head()
```

```python
# List of categorical columns to encode
categorical_cols = ["username", "auth_type", "remote_ip", "lan_ip", "source_os"]
label_encoders = {}  # Dictionary to store encoders for possible reuse

for col in categorical_cols:
    le = LabelEncoder()  # Initialize label encoder
    train_df[col] = le.fit_transform(train_df[col])  # Fit on training data and transform it
    label_encoders[col] = le  # Store the encoder for later use (e.g., inverse_transform)

    # Create a mapping of class → encoded value for consistent test set encoding
    mapping = dict(zip(le.classes_, le.transform(le.classes_)))
    
    # Map test data using training set encoding
    # If unseen category appears in test set, assign -1
    test_df[col] = test_df[col].map(mapping).fillna(-1).astype(int)

print("Categorical encoding completed.")
train_df.head()
```

```python
import joblib  # Add this import if not already done

# Select features for clustering
features = [
    "login_hour",
    "logout_hour",
    "day_of_week",
    "session_duration_seconds",
    "auth_type",
    "remote_ip",
    "lan_ip"
]

# Extract selected features from train and test datasets
X_train = train_df[features].copy()
X_test = test_df[features].copy()

print("Selected features:\n", features)
print("\nSample of unscaled training data:\n", X_train.head())

# Normalize using StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Save the scaler for later use
joblib.dump(scaler, "hdbscan_scaler.pkl")

print("\n Features selected and scaled. Scaler saved. Ready for Training.")
```

```python
from sklearn.decomposition import PCA
from collections import Counter
import joblib

# Reduce dimensionality for visualization
pca = PCA(n_components=2)
X_train_pca = pca.fit_transform(X_train_scaled)

# Define HDBSCAN parameter combinations
params_list = [
    {"min_cluster_size": 20},
    {"min_cluster_size": 30},
    {"min_cluster_size": 40},
    {"min_cluster_size": 30, "min_samples": 10},
    {"min_cluster_size": 40, "min_samples": 20},
    {"min_cluster_size": 50, "min_samples": 10},
    {"min_cluster_size": 50, "min_samples": 20}, 
    {"min_cluster_size": 60, "min_samples": 20},
    {"min_cluster_size": 70, "min_samples": 25},
    {"min_cluster_size": 80, "min_samples": 30},
]

# Plot setup
plt.figure(figsize=(18, 12))

# Tracking best model
best_model = None
best_params = None
lowest_noise = float('inf')

# Run through all parameter combinations
for i, params in enumerate(params_list, 1):
    print(f"\n=== Run {i}: HDBSCAN with params = {params} ===")
    hdb = hdbscan.HDBSCAN(**params, prediction_data=True)
    hdb.fit(X_train_scaled)
    labels = hdb.labels_

    # Cluster and anomaly statistics
    n_clusters = len(set(labels)) - (1 if -1 in labels else 0)
    n_noise = list(labels).count(-1)

    print("Cluster labels found:", sorted(set(labels)))
    print("Label counts:", Counter(labels))
    print("Anomalies detected:", n_noise)
    print("Number of clusters formed:", n_clusters)

    # Save best model (e.g., lowest number of noise points)
    if 200 < n_noise < 5000 and n_clusters > 10 and n_noise < lowest_noise:
        best_model = hdb
        best_params = params
        lowest_noise = n_noise

    # Plot PCA-reduced results
    # plt.subplot(2, 3, i)
    # plt.scatter(
    #     X_train_pca[:, 0],
    #     X_train_pca[:, 1],
    #     c=labels,
    #     cmap="viridis",
    #     s=4,
    #     alpha=0.6
    # )
    # plt.title(f"HDBSCAN: {params}")
    # plt.xlabel("PCA 1")
    # plt.ylabel("PCA 2")

# plt.tight_layout()
# plt.show()

# Save best model
if best_model:
    joblib.dump(best_model, "best_hdbscan_model.pkl")
    print(f"\n✅ Best HDBSCAN model saved with params = {best_params} and {lowest_noise} anomalies.")
else:
    print("\n❌ No suitable HDBSCAN model found.")
```

```python
import joblib
from hdbscan import approximate_predict
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import confusion_matrix, classification_report
import seaborn as sns

# === Step 1: Load the saved scaler and HDBSCAN model
scaler = joblib.load("/kaggle/working/hdbscan_scaler.pkl")
hdb_model = joblib.load("/kaggle/working/best_hdbscan_model.pkl")

# === Step 2: Scale the test data
X_test_scaled = scaler.transform(X_test)

# === Step 3: Predict cluster labels using approximate_predict
predicted_labels, _ = approximate_predict(hdb_model, X_test_scaled)

# === Step 4: Assign cluster labels and mark anomalies (-1 means anomaly)
test_df["cluster_label"] = predicted_labels
test_df["is_anomaly"] = (predicted_labels == -1).astype(int)

# === Step 5: PCA for 2D visualization
pca = PCA(n_components=2)
X_test_pca = pca.fit_transform(X_test_scaled)

# === Plot 1: Cluster View
plt.figure(figsize=(10, 6))
plt.scatter(
    X_test_pca[:, 0],
    X_test_pca[:, 1],
    c=test_df["cluster_label"],
    cmap="viridis",
    s=5,
    alpha=0.7
)
plt.title("HDBSCAN Clusters on Test Data (PCA-Reduced)")
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")
plt.colorbar(label="Cluster Label")
plt.grid(True)
plt.tight_layout()
plt.show()

# === Plot 2: Anomaly Detection View
plt.figure(figsize=(10, 6))
plt.scatter(
    X_test_pca[test_df["is_anomaly"] == 0, 0],
    X_test_pca[test_df["is_anomaly"] == 0, 1],
    c='lightgray',
    s=5,
    label='Normal'
)
plt.scatter(
    X_test_pca[test_df["is_anomaly"] == 1, 0],
    X_test_pca[test_df["is_anomaly"] == 1, 1],
    c='red',
    s=10,
    label='Anomaly'
)
plt.title("Anomalous Sessions in Test Data (PCA-Reduced)")
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# === Step 6: Final Evaluation using Ground Truth
# Replace with your actual ground truth column
y_true = test_labels  # ← this must be test_df["is_anomaly_groundtruth"] or your actual ground truth
y_pred = test_df["is_anomaly"]

print("=== Final HDBSCAN Evaluation ===")
print(classification_report(y_true, y_pred, digits=4))

# === Step 7: Confusion Matrix
cm = confusion_matrix(y_true, y_pred)
labels_matrix = [
    [f"TN = {cm[0][0]}", f"FP = {cm[0][1]}"],
    [f"FN = {cm[1][0]}", f"TP = {cm[1][1]}"]
]

plt.figure(figsize=(6, 4))
sns.heatmap(cm, annot=labels_matrix, fmt="", cmap="Blues",
            xticklabels=["Normal", "Anomaly"],
            yticklabels=["Normal", "Anomaly"])
plt.title("Confusion Matrix: HDBSCAN Final")
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.tight_layout()
plt.show()
```
