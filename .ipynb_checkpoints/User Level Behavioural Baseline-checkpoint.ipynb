```python
# Helper function to aggregate session data
def aggregate_session(df):
    agg = df.groupby(["username", "date"]).agg({
        "session_duration_seconds": ["mean", "sum", "count"],
        "login_time": lambda x: x.dt.hour.mode()[0]
    }).reset_index()
    
    agg.columns = [
        "username", "date", 
        "avg_session_duration", 
        "total_session_duration", 
        "session_count", 
        "most_common_login_hour"
    ]
    return agg

# Apply to train and test
session_agg = aggregate_session(session_df)
session_agg_test = aggregate_session(session_df_test)

# Show samples
print("Aggregated Session Features:\n")
print(session_agg.head())
print("Test Aggregated Session Features:\n")
print(session_agg_test.head())
```

```python
# Helper function to aggregate application usage
def aggregate_app_usage(df):
    agg = df.groupby(["username", "date"]).agg({
        "process_name": ["nunique", "count"],
        "cpu_percent": "mean",
        "memory_percent": "mean"
    }).reset_index()
    
    agg.columns = [
        "username", "date", 
        "unique_apps_used", 
        "total_app_launches", 
        "avg_cpu_percent", 
        "avg_memory_percent"
    ]
    return agg

# Apply to train and test
app_usage_agg = aggregate_app_usage(app_usage_df)
app_usage_agg_test = aggregate_app_usage(app_usage_df_test)

# Show samples
print("Aggregated Application Usage Features:\n")
print(app_usage_agg.head())
print("Test Aggregated Application Usage Features:\n")
print(app_usage_agg_test.head())
```

```python
# Helper function to aggregate executed commands
def aggregate_commands(df):
    command_counts = df.groupby(["username", "date"]).agg({
        "command": ["count", "nunique"]
    }).reset_index()
    command_counts.columns = ["username", "date", "total_commands", "unique_commands"]

    source_pivot = df.groupby(["username", "date", "source"]).size().unstack(fill_value=0).reset_index()

    return pd.merge(command_counts, source_pivot, on=["username", "date"], how="left")

# Apply to train and test
commands_agg = aggregate_commands(commands_df)
commands_agg_test = aggregate_commands(commands_df_test)

# Show samples
print("Aggregated Executed Commands Features:\n")
print(commands_agg.head())
print("Test Aggregated Executed Commands Features:\n")
print(commands_agg_test.head())
```

```python
# Helper function to aggregate resource usage
def aggregate_resource(df):
    agg = df.groupby(["username", "date"]).agg({
        "cpu_percent": "mean",
        "memory_mb": "mean"
    }).reset_index()

    agg.columns = ["username", "date", "avg_daily_cpu", "avg_daily_memory_mb"]
    return agg

# Apply to train and test
resource_agg = aggregate_resource(resource_df)
resource_agg_test = aggregate_resource(resource_df_test)

# Show samples
print("Aggregated Resource Usage Features:\n")
print(resource_agg.head())
print("Test Aggregated Resource Usage Features:\n")
print(resource_agg_test.head())
```

```python
from functools import reduce

# Helper function to merge list of dataframes on username and date
def merge_features(dfs):
    merged = reduce(lambda left, right: pd.merge(left, right, on=["username", "date"], how="outer"), dfs)
    return merged.sort_values(by=["username", "date"])

# Merge train and test features
baseline_df = merge_features([session_agg, app_usage_agg, commands_agg, resource_agg])
baseline_df_test = merge_features([session_agg_test, app_usage_agg_test, commands_agg_test, resource_agg_test])

# Show samples
print("Unified Baseline Feature Set:\n")
print(baseline_df.head())
print("Unified Test Feature Set:\n")
print(baseline_df_test.head())
```

```python
# === Step: Aggregate per-user-per-day anomalies from original test datasets ===
# Convert is_anomaly to boolean if needed
for df in [session_df_test, app_usage_df_test, commands_df_test, resource_df_test]:
    df["is_anomaly"] = df["is_anomaly"].astype(bool)

# Per-user-per-day boolean flags from each test dataset
session_anom = session_df_test.groupby(["username", "date"])["is_anomaly"].any().reset_index().rename(columns={"is_anomaly": "session_anomaly"})
app_anom     = app_usage_df_test.groupby(["username", "date"])["is_anomaly"].any().reset_index().rename(columns={"is_anomaly": "app_anomaly"})
cmd_anom     = commands_df_test.groupby(["username", "date"])["is_anomaly"].any().reset_index().rename(columns={"is_anomaly": "cmd_anomaly"})
res_anom     = resource_df_test.groupby(["username", "date"])["is_anomaly"].any().reset_index().rename(columns={"is_anomaly": "res_anomaly"})

# Merge all flags into one dataframe
from functools import reduce
df_agg = reduce(lambda left, right: pd.merge(left, right, on=["username", "date"], how="outer"), [session_anom, app_anom, cmd_anom, res_anom])
df_agg.fillna(False, inplace=True)  # fill missing with False

# === Step: Compute weighted anomaly score ===
weights = {
    "session_anomaly": 0.2,
    "app_anomaly": 0.2,
    "cmd_anomaly": 0.3,
    "res_anomaly": 0.1
}

df_agg["weighted_score"] = (
    df_agg["session_anomaly"].astype(float) * weights["session_anomaly"] +
    df_agg["app_anomaly"].astype(float)     * weights["app_anomaly"] +
    df_agg["cmd_anomaly"].astype(float)     * weights["cmd_anomaly"] +
    df_agg["res_anomaly"].astype(float)     * weights["res_anomaly"]
)

# Flag true anomaly based on threshold
threshold = 0.5  # tuneable
df_agg["true_anomaly"] = (df_agg["weighted_score"] >= threshold).astype(int)

# Preview
print("Per-user-per-day Ground Truth Anomalies:\n")
print(df_agg.head())
```

```python
from sklearn.preprocessing import StandardScaler

# Helper function to scale features and reattach identifiers
def scale_features(df, feature_cols, scaler=None, fit=True):
    X = df[feature_cols].copy()
    if fit:
        scaler = StandardScaler()
        X_scaled = scaler.fit_transform(X)
    else:
        X_scaled = scaler.transform(X)
    scaled_df = pd.DataFrame(X_scaled, columns=feature_cols)
    scaled_df["username"] = df["username"].values
    scaled_df["date"] = df["date"].values
    return scaled_df, scaler

# Identify feature columns (excluding identifiers)
feature_cols = baseline_df.columns.difference(["username", "date"])

# Scale train and test
X_scaled_df, scaler = scale_features(baseline_df, feature_cols, fit=True)
X_test_scaled_df, _ = scale_features(baseline_df_test, feature_cols, scaler=scaler, fit=False)

# Show samples
print("Scaled feature matrix shape:", X_scaled_df.shape)
print(X_scaled_df.head())
print("Scaled Test Feature Matrix:\n", X_test_scaled_df.shape)
print(X_test_scaled_df.head())
```

```python
from sklearn.ensemble import IsolationForest
from itertools import product
from tqdm import tqdm
import numpy as np
import logging
import joblib
import os

# Configure logging
logging.basicConfig(level=logging.WARNING, format='%(levelname)s: %(message)s')

# === Set sliding window size (tunable) ===
k = 3  # changeable window size

# === Prepare feature columns ===
global_feature_cols = [col for col in X_scaled_df.columns if col not in ["username", "date"]]
print("Feature columns used:", sorted(global_feature_cols))

# === Build sliding windows ===
def generate_sliding_windows(df, k):
    sequences = []
    df_sorted = df.sort_values(by=["username", "date"])
    for user, group in df_sorted.groupby("username"):
        X = group[global_feature_cols].values
        if len(X) < k:
            continue
        for i in range(len(X) - k + 1):
            window = X[i:i+k].flatten()
            sequences.append(window)
    return np.array(sequences)

X_global_train = generate_sliding_windows(X_scaled_df, k)
print("Sliding window training shape:", X_global_train.shape)

# === Define hyperparameter grid ===
param_grid = {
    "n_estimators": [100, 200, 300],
    "max_samples": ["auto", 0.8, 0.6],
    "contamination": [0.01, 0.02, 0.03, 0.04],
    "max_features": [1.0, 0.8, 0.6],
    "bootstrap": [False, True],
    "random_state": [42]
}
param_combos = list(product(*param_grid.values()))
param_names = list(param_grid.keys())

# === Model selection ===
best_score = -np.inf
best_global_iforest = None

for combo in tqdm(param_combos, desc="Training Global Isolation Forest (Sliding Window)"):
    params = dict(zip(param_names, combo))
    try:
        model = IsolationForest(**params)
        model.fit(X_global_train)
        scores = model.decision_function(X_global_train)
        score_spread = np.std(scores)

        if score_spread > best_score:
            best_score = score_spread
            best_global_iforest = model

    except Exception as e:
        logging.warning(f"Global model failed with params {params} due to {e}")
        continue

# === Final confirmation ===
if best_global_iforest is not None:
    print(" Best Global Isolation Forest model trained (Sliding Window).")
    print("Best score spread:", best_score)
else:
    print(" Failed to train a global Isolation Forest model.")

# === Save the model ===
model_dir = "./models"
os.makedirs(model_dir, exist_ok=True)
model_path = os.path.join(model_dir, "global_isolation_forest_k3.joblib")
joblib.dump(best_global_iforest, model_path)
print(f"Model saved to: {model_path}")
```

```python
# 1. Histogram of Decision Scores (Training Set)
# Shows the distribution of how "normal" or "abnormal" the training samples are.

import matplotlib.pyplot as plt

# Get decision scores (higher = more normal)
train_scores = best_global_iforest.decision_function(X_global_train)

plt.figure(figsize=(8, 4))
plt.hist(train_scores, bins=50, color='skyblue', edgecolor='black')
plt.title("Isolation Forest Decision Scores (Training Set)")
plt.xlabel("Anomaly Score (higher = more normal)")
plt.ylabel("Count")
plt.tight_layout()
plt.show()

# 2. Threshold Visualization
# You can visualize how many training points fall below a chosen anomaly threshold:

threshold = np.percentile(train_scores, 1)  # bottom 5% as anomalies
anomalies = train_scores < threshold

plt.figure(figsize=(10, 4))
plt.plot(train_scores, marker='o', linestyle='', alpha=0.6, label="Samples")
plt.axhline(threshold, color='red', linestyle='--', label=f"Threshold ({threshold:.4f})")
plt.title("Decision Function Scores with Threshold")
plt.xlabel("Sample Index (sliding window)")
plt.ylabel("Score")
plt.legend()
plt.tight_layout()
plt.show()

#  3. Dimensionality Reduction (PCA)
# Shows how the sliding-windowed input is distributed in 2D space and highlights model separation.
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
X_reduced = pca.fit_transform(X_global_train)

plt.figure(figsize=(8, 6))
plt.scatter(X_reduced[:, 0], X_reduced[:, 1], c=train_scores, cmap='viridis', s=20)
plt.colorbar(label="Anomaly Score (higher = more normal)")
plt.title("PCA of Sliding Window Input Colored by Decision Score")
plt.xlabel("PC 1")
plt.ylabel("PC 2")
plt.tight_layout()
plt.show()
```

```python
# === Predict on test set with sliding windows and evaluate ===
import numpy as np
import pandas as pd
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score

# Use the same k and feature columns as training
# k must match the value used during training cell
# global_feature_cols must already be defined
print(f"Using sliding window size k={k}")
print("Evaluating features:", sorted(global_feature_cols))

def generate_sliding_windows_with_meta(df, k, feature_cols):
    """
    Build rolling windows per user and return:
    - X_windows: flattened window vectors (n_samples, k * n_features)
    - users: user for each window end
    - end_dates: date corresponding to the last day in the window
    """
    seqs, users, ends = [], [], []
    df_sorted = df.sort_values(by=["username", "date"])
    for user, group in df_sorted.groupby("username"):
        X = group[feature_cols].values
        dates = pd.to_datetime(group["date"]).values
        if len(X) < k:
            continue
        for i in range(len(X) - k + 1):
            seqs.append(X[i:i+k].flatten())
            users.append(user)
            ends.append(dates[i + k - 1])
    if len(seqs) == 0:
        return np.empty((0, k * len(feature_cols))), np.array([]), np.array([])
    return np.array(seqs), np.array(users), pd.to_datetime(ends)

# Build test windows
X_test_windows, users_test, end_dates_test = generate_sliding_windows_with_meta(
    X_test_scaled_df, k, global_feature_cols
)
print("Test windows shape:", X_test_windows.shape)

# Sanity checks
if X_test_windows.shape[0] == 0:
    raise RuntimeError("No test windows generated. Check k and test data availability.")

# Predict with the best global Isolation Forest
scores = best_global_iforest.decision_function(X_test_windows)  # higher => more normal
pred_raw = best_global_iforest.predict(X_test_windows)          # 1=normal, -1=anomaly
pred_anom = (pred_raw == -1).astype(int)                        # 1=anomaly

# Build predictions DataFrame aligned to (username, date=end_of_window)
preds_df = pd.DataFrame({
    "username": users_test,
    "date": pd.to_datetime(end_dates_test),
    "iforest_score": scores,
    "predicted_anomaly": pred_anom
})

# Merge with ground truth (may be NaN for days without labels or before k)
gt_df = baseline_df_test[["username", "date", "true_anomaly"]].copy()
gt_df["date"] = pd.to_datetime(gt_df["date"])

eval_df = gt_df.merge(preds_df, on=["username", "date"], how="left")

# Keep only rows where we produced a prediction (i.e., after warm-up window)
eval_mask = eval_df["predicted_anomaly"].notna()
eval_eval = eval_df.loc[eval_mask].copy()
eval_eval["predicted_anomaly"] = eval_eval["predicted_anomaly"].astype(int)

# Metrics
y_true = eval_eval["true_anomaly"].astype(int).values
y_pred = eval_eval["predicted_anomaly"].astype(int).values
y_score = (-eval_eval["iforest_score"].values)  # negate because higher score = more normal

cm = confusion_matrix(y_true, y_pred, labels=[0,1])
report = classification_report(y_true, y_pred, digits=4, target_names=["Normal(0)", "Anomaly(1)"], zero_division=0)
try:
    roc_auc = roc_auc_score(y_true, y_score)
except ValueError:
    roc_auc = np.nan

print("\n=== Evaluation (on days with window-based predictions) ===")
print(f"Coverage: {eval_eval.shape[0]} predicted day-windows out of {gt_df.shape[0]} total test rows")
print("Confusion Matrix [[TN, FP],[FN, TP]]:\n", cm)
print("\nClassification Report:\n", report)
print("ROC-AUC (decision scores):", roc_auc)

# Optional: attach predictions back to baseline_df_test (keep NaN for days without predictions)
baseline_df_test = baseline_df_test.merge(
    preds_df, on=["username", "date"], how="left"
)

print("\nSample predictions merged into baseline_df_test:\n",
      baseline_df_test[["username", "date", "true_anomaly", "predicted_anomaly", "iforest_score"]].head(10))
```
