```python
# === Cell 1: Import Libraries ===

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from datetime import datetime, timedelta
import os
import warnings

warnings.filterwarnings("ignore")
```

```python
# === Cell 2: Load Training and Testing Datasets ===

# Define base path
base_path = "/home/simar/Documents/UEBA_AI/Anomalous CPU consumption/datasets"

# Load CSVs
train_df = pd.read_csv(f"{base_path}/ueba_training_dataset_500_users.csv", parse_dates=["login_time", "logout_time"])
# test_df = pd.read_csv(f"{base_path}/ueba_testing_dataset_500_users.csv", parse_dates=["login_time", "logout_time"])
test_df = pd.read_csv(f"{base_path}/ueba_testing_dataset_500_users_labeled.csv", parse_dates=["login_time", "logout_time"])

# Quick shape check
print("Training set shape:", train_df.shape)
print("Testing set shape:", test_df.shape)
```

```python
# === Cell 3: Data Overview and Visualizations ===

# CPU usage distribution (train vs test)
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.histplot(train_df["cpu_usage"], bins=40, kde=True, color='skyblue')
plt.title("CPU Usage Distribution (Training Set)")
plt.xlabel("CPU Usage (%)")

plt.subplot(1, 2, 2)
sns.histplot(test_df["cpu_usage"], bins=40, kde=True, color='salmon')
plt.title("CPU Usage Distribution (Testing Set)")
plt.xlabel("CPU Usage (%)")

plt.tight_layout()
plt.show()

# Boxplot for spotting outliers
plt.figure(figsize=(8, 4))
sns.boxplot(data=train_df, x="cpu_usage", color='lightgreen')
plt.title("Boxplot of CPU Usage (Training Set)")
plt.xlabel("CPU Usage (%)")
plt.tight_layout()
plt.show()
```

```python
# === Cell 4: Ground Truth Anomaly Overview (Test Set) ===

# Total and anomaly counts
total = len(test_df)
anomalies = test_df["is_anomaly"].sum()
normals = total - anomalies
anomaly_percent = (anomalies / total) * 100

print(f"Total test records: {total}")
print(f"Anomalies: {anomalies} ({anomaly_percent:.2f}%)")
print(f"Normal records: {normals}")
```

```python
# === Cell 4: Add Date Column for Daily Aggregation ===

train_df["date"] = train_df["login_time"].dt.date
test_df["date"] = test_df["login_time"].dt.date

# Confirm changes
print("Sample dates (training):", train_df["date"].unique()[:3])
print("Sample dates (testing):", test_df["date"].unique()[:3])
```

```python
# === Cell 5: Aggregate Per-User-Per-Day CPU Usage ===

# Training aggregation
cpu_agg_train = train_df.groupby(["username", "date"]).agg({
    "cpu_usage": "mean"
}).reset_index()
cpu_agg_train.rename(columns={"cpu_usage": "avg_daily_cpu"}, inplace=True)

# Testing aggregation
cpu_agg_test = test_df.groupby(["username", "date"]).agg({
    "cpu_usage": "mean"
}).reset_index()
cpu_agg_test.rename(columns={"cpu_usage": "avg_daily_cpu"}, inplace=True)

# Preview
print("Aggregated Training Data:\n", cpu_agg_train.head())
print("\nAggregated Testing Data:\n", cpu_agg_test.head())
```

```python
# === Cell 6: Feature Scaling ===

from sklearn.preprocessing import StandardScaler

# Columns to scale
feature_cols = ["avg_daily_cpu"]

# Initialize scaler and fit on training data
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(cpu_agg_train[feature_cols])
X_test_scaled = scaler.transform(cpu_agg_test[feature_cols])

# Reattach user and date info
X_train_df = pd.DataFrame(X_train_scaled, columns=feature_cols)
X_train_df["username"] = cpu_agg_train["username"].values
X_train_df["date"] = cpu_agg_train["date"].values

X_test_df = pd.DataFrame(X_test_scaled, columns=feature_cols)
X_test_df["username"] = cpu_agg_test["username"].values
X_test_df["date"] = cpu_agg_test["date"].values

# Preview
print("Scaled Training Data:\n", X_train_df.head())
print("\nScaled Testing Data:\n", X_test_df.head())
```

```python
# === Cell 7: Sliding Window Preparation (Global Model) ===

import numpy as np

# Set sliding window size
k = 3  # You can change this later if needed

# Define feature columns (only avg_daily_cpu for now)
global_feature_cols = ["avg_daily_cpu"]

def generate_sliding_windows(df, k):
    sequences = []
    
    # Sort by username and date
    df_sorted = df.sort_values(by=["username", "date"])
    
    for user, group in df_sorted.groupby("username"):
        X = group[global_feature_cols].values
        if len(X) < k:
            continue
        for i in range(len(X) - k + 1):
            window = X[i:i+k].flatten()
            sequences.append(window)
    
    return np.array(sequences)

# Build global training window set
X_global_train = generate_sliding_windows(X_train_df, k)
print("Global sliding window training shape:", X_global_train.shape)
```

```python
# === Cell 8: Train Global Isolation Forest (Grid Search + Save Best Model) ===

from sklearn.ensemble import IsolationForest
from itertools import product
from tqdm import tqdm
import numpy as np
import logging
import joblib
import os

# Configure logging
logging.basicConfig(level=logging.WARNING, format='%(levelname)s: %(message)s')

# === Define hyperparameter grid ===
param_grid = {
    "n_estimators": [100, 200,300,400,500],
    "max_samples": [0.4, 0.5, 0.6],
    "contamination": [0.1, 0.2, 0.3, 0.4],
    "max_features": [1.0, 0.8, 0.6],
    "bootstrap": [False,True],
    "random_state": [42]
}
param_combos = list(product(*param_grid.values()))
param_names = list(param_grid.keys())

# === Model selection ===
best_score = -np.inf
best_global_iforest = None

for combo in tqdm(param_combos, desc="Training Global Isolation Forest (Sliding Window)"):
    params = dict(zip(param_names, combo))
    try:
        model = IsolationForest(**params)
        model.fit(X_global_train)
        scores = model.decision_function(X_global_train)
        # score_spread = np.std(scores)
        score_spread = np.percentile(scores, 95) - np.percentile(scores, 5)


        if score_spread > best_score:
            best_score = score_spread
            best_global_iforest = model

    except Exception as e:
        logging.warning(f"Model failed with params {params} due to {e}")
        continue

# === Final confirmation ===
if best_global_iforest is not None:
    print(" Best Global Isolation Forest model trained (Sliding Window).")
    print("Best score spread:", best_score)
else:
    print(" Failed to train a global Isolation Forest model.")

# === Save the best model ===
model_dir = "./models"
os.makedirs(model_dir, exist_ok=True)
model_path = os.path.join(model_dir, "global_isolation_forest_k3_cpu_only.joblib")
joblib.dump(best_global_iforest, model_path)
print(f"Model saved to: {model_path}")
```

```python
# === Cell 9: Model Score Visualization ===

import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

# Get decision scores from trained model
train_scores = best_global_iforest.decision_function(X_global_train)

# --- 1. Histogram of Decision Scores ---
plt.figure(figsize=(8, 4))
plt.hist(train_scores, bins=50, color='skyblue', edgecolor='black')
plt.title("Isolation Forest Decision Scores (Training Set)")
plt.xlabel("Anomaly Score (higher = more normal)")
plt.ylabel("Count")
plt.tight_layout()
plt.show()

# --- 2. Threshold Visualization ---
threshold = np.percentile(train_scores, 1)  # bottom 1% as anomalies
anomalies = train_scores < threshold

plt.figure(figsize=(10, 4))
plt.plot(train_scores, marker='o', linestyle='', alpha=0.6, label="Samples")
plt.axhline(threshold, color='red', linestyle='--', label=f"Threshold ({threshold:.4f})")
plt.title("Decision Scores with Anomaly Threshold")
plt.xlabel("Sample Index")
plt.ylabel("Score")
plt.legend()
plt.tight_layout()
plt.show()

# --- 3. PCA Projection ---
pca = PCA(n_components=2)
X_reduced = pca.fit_transform(X_global_train)

plt.figure(figsize=(8, 6))
plt.scatter(X_reduced[:, 0], X_reduced[:, 1], c=train_scores, cmap='viridis', s=20)
plt.colorbar(label="Anomaly Score (higher = more normal)")
plt.title("PCA of Sliding Windows Colored by Decision Score")
plt.xlabel("PC 1")
plt.ylabel("PC 2")
plt.tight_layout()
plt.show()
```

```python
# === Cell 10: Load Trained Global Model and Prepare Test Sliding Windows ===

import joblib

# Load saved model
model_path = "./models/global_isolation_forest_k3_cpu_only.joblib"
loaded_global_iforest = joblib.load(model_path)
print("Model loaded from:", model_path)

# Build sliding windows on test set
def generate_sliding_windows_with_meta(df, k, feature_cols):
    """
    Build rolling windows per user and return:
    - X_windows: flattened window vectors (n_samples, k * n_features)
    - users: user for each window end
    - end_dates: date corresponding to the last day in the window
    """
    seqs, users, ends = [], [], []
    df_sorted = df.sort_values(by=["username", "date"])
    for user, group in df_sorted.groupby("username"):
        X = group[feature_cols].values
        dates = pd.to_datetime(group["date"]).values
        if len(X) < k:
            continue
        for i in range(len(X) - k + 1):
            seqs.append(X[i:i+k].flatten())
            users.append(user)
            ends.append(dates[i + k - 1])
    return np.array(seqs), np.array(users), pd.to_datetime(ends)

# Generate test windows
X_test_windows, users_test, end_dates_test = generate_sliding_windows_with_meta(
    X_test_df, k, global_feature_cols
)

print("Test windows shape:", X_test_windows.shape)
```

```python
# === Cell 11: Predict Anomalies on Test Sliding Windows (with Confidence Scores) ===

from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score
from sklearn.preprocessing import MinMaxScaler

# Predict decision scores from the model
test_scores = loaded_global_iforest.decision_function(X_test_windows)  # Higher = more normal

# Normalize to get anomaly confidence (0 = normal, 1 = most anomalous)
scaler = MinMaxScaler()
normalized = scaler.fit_transform(test_scores.reshape(-1, 1)).flatten()
anomaly_confidence = 1 - normalized  # Flip: higher = more anomalous

# Optionally, set threshold for binary classification
threshold = 0.80
predicted_anomaly = (anomaly_confidence > threshold).astype(int)

# Create predictions DataFrame
preds_df = pd.DataFrame({
    "username": users_test,
    "date": end_dates_test,
    "iforest_score": test_scores,
    "anomaly_confidence": anomaly_confidence,
    "predicted_anomaly": predicted_anomaly
})

# Preview
print("Sample predictions:\n", preds_df.head())
print("Anomaly count:", preds_df["predicted_anomaly"].sum())
```

```python
# === Cell: Evaluate Model Performance Against Ground Truth Labels ===

from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, ConfusionMatrixDisplay
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Reload the labeled test set to ensure it has 'is_anomaly' and 'date'
test_df = pd.read_csv("/home/simar/Documents/UEBA_AI/Anomalous CPU consumption/datasets/ueba_testing_dataset_500_users_labeled.csv", parse_dates=["login_time"])
test_df["date"] = test_df["login_time"].dt.normalize()

# Normalize date in preds_df
preds_df["date"] = pd.to_datetime(preds_df["date"]).dt.normalize()

# Merge predictions with ground truth
evaluation_df = preds_df.merge(
    test_df[["username", "date", "is_anomaly"]],
    on=["username", "date"],
    how="left"
)

# Drop rows without ground truth
evaluation_df = evaluation_df.dropna(subset=["is_anomaly"])
evaluation_df["is_anomaly"] = evaluation_df["is_anomaly"].astype(int)

# Confusion matrix
cm = confusion_matrix(evaluation_df["is_anomaly"], evaluation_df["predicted_anomaly"])
print("Confusion Matrix [[TN, FP], [FN, TP]]:\n", cm)

# Classification report
report = classification_report(
    evaluation_df["is_anomaly"],
    evaluation_df["predicted_anomaly"],
    target_names=["Normal (0)", "Anomaly (1)"]
)
print("\nClassification Report:\n", report)

# ROC AUC
roc_auc = roc_auc_score(evaluation_df["is_anomaly"], evaluation_df["iforest_score"])
print("ROC-AUC (decision function score):", round(roc_auc, 4))

# === Visualizations ===

# Confusion Matrix Plot
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["Normal", "Anomaly"])
disp.plot(cmap="Blues")
plt.title("Confusion Matrix")
plt.grid(False)
plt.tight_layout()
plt.show()

# Anomaly score distributions
plt.figure(figsize=(8, 5))
sns.histplot(data=evaluation_df, x="iforest_score", hue="is_anomaly", bins=50, palette={0: "skyblue", 1: "salmon"}, kde=True)
plt.title("Isolation Forest Decision Scores by Ground Truth")
plt.xlabel("Anomaly Score (higher = more normal)")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```
